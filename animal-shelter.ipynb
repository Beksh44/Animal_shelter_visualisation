## age_to_months():
This function takes the age as a string (e.g., "2 years", "5 months") and converts it into a numerical value in months. It handles multiple units such as "years", "months", "weeks", and "days".  
1 year = 12 months  
1 week ≈ 0.25 months  
1 day ≈ 0.033 months*
1. Check if the age string is null
2. Extract the numeric part from the age string.
3. Identify the unit (years, months, weeks, days).
4. Convert all values to a single unit (months).
  
Then we convert Age upon Intake and Age upon Outcome into a numerical format(float64, because it can store NaN values)

```python
def age_to_days(age_str):
    if pd.isnull(age_str):
        return np.nan
    age_parts = age_str.split()
    if len(age_parts) == 2:
        num, unit = int(age_parts[0]), age_parts[1].lower()
        if 'year' in unit:
            return num * 365 if num >= 0 else np.nan
        elif 'month' in unit:
            return num * 30 if num >= 0 else np.nan
        elif 'week' in unit:
            return num * 7 if num >= 0 else np.nan
        elif 'day' in unit:
            return num if num >= 0 else np.nan
    return np.nan

# Convert 'Age upon Intake' in intakes_df
intakes_df['Age upon Intake'] = intakes_df['Age upon Intake'].apply(age_to_days)

# Convert 'Age upon Outcome' in outcomes_df
outcomes_df['Age upon Outcome'] = outcomes_df['Age upon Outcome'].apply(age_to_days)

# Drop rows with NaN values before converting to int64
intakes_df['Age upon Intake'] = intakes_df['Age upon Intake'].dropna().astype('Int64')
outcomes_df['Age upon Outcome'] = outcomes_df['Age upon Outcome'].dropna().astype('Int64')
```

```python
# Count the number of duplicate rows in the DataFrame
def count_duplicate(df):
    num_duplicates = df.duplicated().sum()

    if num_duplicates == 0:
        print("No duplicates found.")
    else:
        print(f"There are {num_duplicates} duplicate rows.")

count_duplicate(intakes_df)
count_duplicate(outcomes_df)
```
